## Regular Expresssion
- Set of rules or set of metacharacters.

```python
import re
```

```python
# Simple example on RegEx
# Display the lines which are containing "how"

fhand = open("myfile.txt")
for line in fhand:
    line = line.rstrip() #remove characters from the right side
    if re.search("how",line): #re.search does pattern search. First argument gives pattern to be searched and 2nd one gives from where should it search.
        print(line)
        
fhand.close()
```

```python
fhand = open("myfile.txt")
for line in fhand:
    line = line.rstrip() #remove characters from the right side
    if re.search("^[hH]ow",line):  # Matches beginning of the line
        print(line)
fhand.close()
```

```python
fhand = open("myfile.txt")
for line in fhand:
    line = line.rstrip() #remove characters from the right side
    if re.search("^I..m",line):  # Prints any line starting with I and after I has 2 characters(..) and then has m. (ie two characters between I and m)
        print(line)
fhand.close() 
# Here output "I" "space" "a" "m"
```

```python
fhand = open("myfile.txt")
for line in fhand:
    line = line.rstrip() 
    if re.search("^h.+u",line):
        print(line)
# here (.) means any character,so (.+) means any charater may appear one or more times.THEREFORE it means starts with h and has n number of charaters and then has u.
fhand.close() 
```

```python
fhand = open("myfile.txt")
for line in fhand:
    line = line.rstrip() 
    if re.search("^h.*u",line):
        print(line)
# h then zero or more charaters then u. ie hu will also be displayed
fhand.close() 
```

```python
'''
Write a program to extract lines starting with the word "From" (or "from) and ending with .edu
Input file is: mbox-short.txt
'''
fhand = open("mbox-short.txt")
count = 0
for line in fhand:
    count +=1
    line = line.strip()
    pattern ="^[fF]rom.*edu$"
    if re.search(pattern,line):
        print("line number: ",count,"Line: ",line)
fhand.close()
```

```python
# Extract all the line ending with any digit
fhand = open("mbox-short.txt")
for line in fhand:
    line = line.strip()
    pattern = "[0-9]$"
    if re.search(pattern,line):
        print("Line: ",line)
fhand.close()
```

```python
# line must be starting with anything other than lowercase alphabets and digits
fhand=open("mbox-short.txt")
for line in fhand:
    line = line.strip()
    pattern = "^[^a-z0-9]" # It will automatically distinguish between a- z and 0-9, NO NEED OF "-" OR SPACE
    if re.search(pattern,line):
        print("Line: ",line)
fhand.close()
```

```python
# Line should start with capital letters and end with any digit
fhand =open("mbox-short.txt")
for line in fhand:
    line = line.strip()
    pattern = "^[A-Z].*[0-9]$" #or ^[A-Z].*\d$
    if re.search(pattern,line):
        print("Line: ", line )
fhand.close()
```

```python
import re
s = "A message from csev@umich.edu to cwen@iupui.edu about meeting @2PM "

lst  = re.findall("\S+@\S+",s)
print(lst)

#Extract only the domain names
for email in lst:
    x = email.find("@")
    print(email[x+1:])
```

```python
# Extract all the email IDs from the file mbox-short.txt
fhand =open("mbox-short.txt")
for line in fhand:
    line = line.strip()
    pattern = "\S+@\S+"
    lst = re.findall("\S+@\S+",line)
    if len(lst)>0:
        print(lst)
fhand.close()

# Here some additional charaters are displayed along with the email id (ie. ',<,; etc) and also "apache@localhost" is not a email id.
```

```python
#so
## Extract all the email IDs from the file mbox-short.txt
fhand = open("mbox-short.txt")
ls = []
pattern = "[a-zA-Z0-9]\S+@\S+[a-zA-Z][.][a-zA-Z]+" # [a-zA-Z0-9]\S+@\S+[.][a-zA-Z]+
for line in fhand:
    line = line.strip()
    x = re.findall(pattern,line)
    if len(x)>0:
        ls.extend(x)
        
print(ls)
```
